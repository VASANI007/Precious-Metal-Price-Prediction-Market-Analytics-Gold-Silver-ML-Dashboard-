```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/processed/final_data.csv")
df['Date'] = pd.to_datetime(df['Date'])

df.head()
```

```python
df.info()
df.describe()
```

```python
# ensure datetime
df['Date'] = pd.to_datetime(df['Date'])
df = df.copy()
plt.style.use('dark_background')
plt.figure(figsize=(14,6))
# main line
plt.plot(
    df['Date'], 
    df['Gold_24K_1g'], 
    color='#FFD700',  # gold color
    linewidth=2.5, 
    label='Gold Price (1g)'
)
# markers
plt.scatter(
    df['Date'].iloc[::50],  
    df['Gold_24K_1g'].iloc[::50],
    color='white',
    s=25,
    alpha=0.8,
    label='Data Points'
)
# moving average
df.loc[:, 'MA_30'] = df['Gold_24K_1g'].rolling(window=30).mean()
plt.plot(
    df['Date'],
    df['MA_30'],
    color='#FF4C4C', 
    linestyle='--',
    linewidth=2,
    label='30-Day Avg'
)
# title & labels
plt.title("Gold 24K Price Trend (1g)", fontsize=16, fontweight='bold', color='white')
plt.xlabel("Date", fontsize=12, color='white')
plt.ylabel("Price (INR)", fontsize=12, color='white')
# grey grid
plt.grid(True, linestyle='--', linewidth=0.5, color='gray', alpha=0.4)
# ticks color
plt.xticks(color='white')
plt.yticks(color='white')
# legend
plt.legend(facecolor='black', edgecolor='white')
# spacing
plt.tight_layout()
plt.show()
```

```python
#Silver Price Trend

# ensure datetime
df['Date'] = pd.to_datetime(df['Date'])
df = df.copy()

# dark theme
plt.style.use('dark_background')
plt.figure(figsize=(14,6))
# main silver line
plt.plot(
    df['Date'], 
    df['Silver_1g'], 
    color='#C0C0C0',  # silver color
    linewidth=2.5, 
    label='Silver Price (1g)'
)
# markers
plt.scatter(
    df['Date'].iloc[::50],  
    df['Silver_1g'].iloc[::50],
    color='white',
    s=25,
    alpha=0.8,
    label='Data Points'
)
# moving average
df.loc[:, 'MA_30_Silver'] = df['Silver_1g'].rolling(window=30).mean()
plt.plot(
    df['Date'],
    df['MA_30_Silver'],
    color='#00BFFF',  # light blue
    linestyle='--',
    linewidth=2,
    label='30-Day Avg'
)
# title & labels
plt.title("Silver Price Trend (1g)", fontsize=16, fontweight='bold', color='white')
plt.xlabel("Date", fontsize=12, color='white')
plt.ylabel("Price (INR)", fontsize=12, color='white')
# grey grid
plt.grid(True, linestyle='--', linewidth=0.5, color='gray', alpha=0.4)
# ticks
plt.xticks(color='white')
plt.yticks(color='white')
# legend
plt.legend(facecolor='black', edgecolor='white')
# spacing
plt.tight_layout()
plt.show()
```

```python
#Gold vs Silver Comparison
# ensure datetime
df['Date'] = pd.to_datetime(df['Date'])
df = df.copy()

# dark theme
plt.style.use('dark_background')

plt.figure(figsize=(14,6))

# gold line
plt.plot(
    df['Date'], 
    df['Gold_24K_1g'], 
    color='#FFD700', 
    linewidth=2.5, 
    label='Gold (1g)'
)

# silver line
plt.plot(
    df['Date'], 
    df['Silver_1g'], 
    color='#C0C0C0', 
    linewidth=2.5, 
    label='Silver (1g)'
)

# moving averages
df.loc[:, 'Gold_MA'] = df['Gold_24K_1g'].rolling(30).mean()
df.loc[:, 'Silver_MA'] = df['Silver_1g'].rolling(30).mean()

plt.plot(df['Date'], df['Gold_MA'], color='#FF4C4C', linestyle='--', linewidth=2, label='Gold Avg')
plt.plot(df['Date'], df['Silver_MA'], color='#00BFFF', linestyle='--', linewidth=2, label='Silver Avg')

# markers (optional highlight)
plt.scatter(df['Date'].iloc[::80], df['Gold_24K_1g'].iloc[::80], color='white', s=15)
plt.scatter(df['Date'].iloc[::80], df['Silver_1g'].iloc[::80], color='lightblue', s=15)

# title & labels
plt.title("Gold vs Silver Price Comparison", fontsize=16, fontweight='bold', color='white')
plt.xlabel("Date", fontsize=12, color='white')
plt.ylabel("Price (INR)", fontsize=12, color='white')

# grid (grey)
plt.grid(True, linestyle='--', linewidth=0.5, color='gray', alpha=0.4)

# ticks
plt.xticks(color='white')
plt.yticks(color='white')

# legend
plt.legend(facecolor='black', edgecolor='white')

# spacing
plt.tight_layout()
plt.show()
```

```python
#Moving Average

# ensure datetime
df['Date'] = pd.to_datetime(df['Date'])
df = df.copy()

# moving average
df.loc[:, 'MA_30'] = df['Gold_24K_1g'].rolling(window=30).mean()

# dark theme
plt.style.use('dark_background')

plt.figure(figsize=(14,6))

# original line
plt.plot(
    df['Date'], 
    df['Gold_24K_1g'], 
    color='#FFD700', 
    linewidth=2, 
    alpha=0.7,
    label="Original Price"
)

# moving average line
plt.plot(
    df['Date'], 
    df['MA_30'], 
    color='#FF4C4C', 
    linestyle='--', 
    linewidth=2.5,
    label="30-Day Moving Avg"
)

# area fill (premium look)
plt.fill_between(
    df['Date'], 
    df['Gold_24K_1g'], 
    color='#FFD700', 
    alpha=0.08
)

# highlight crossover points
plt.scatter(
    df['Date'].iloc[::80],
    df['Gold_24K_1g'].iloc[::80],
    color='white',
    s=20
)

# title & labels
plt.title("Gold Price with Moving Average", fontsize=16, fontweight='bold', color='white')
plt.xlabel("Date", fontsize=12, color='white')
plt.ylabel("Price (INR)", fontsize=12, color='white')

# grid (grey)
plt.grid(True, linestyle='--', linewidth=0.5, color='gray', alpha=0.4)

# ticks
plt.xticks(color='white')
plt.yticks(color='white')

# legend
plt.legend(facecolor='black', edgecolor='white')

# spacing
plt.tight_layout()

plt.show()
```

```python
#Year-wise Analysis

# ensure datetime
df['Date'] = pd.to_datetime(df['Date'])
df = df.copy()

# extract year
df['Year'] = df['Date'].dt.year

# yearly average
yearly_avg = df.groupby('Year')['Gold_24K_1g'].mean()

# dark theme
plt.style.use('dark_background')

plt.figure(figsize=(12,6))

# bar chart
bars = plt.bar(
    yearly_avg.index, 
    yearly_avg.values, 
    color='#FFD700', 
    edgecolor='white'
)

# add value labels on bars
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height,
        f'{height:.0f}',
        ha='center',
        va='bottom',
        fontsize=9,
        color='white'
    )

# title & labels
plt.title("Yearly Average Gold Price", fontsize=16, fontweight='bold', color='white')
plt.xlabel("Year", fontsize=12, color='white')
plt.ylabel("Price (INR)", fontsize=12, color='white')

# grid (grey horizontal only)
plt.grid(axis='y', linestyle='--', linewidth=0.5, color='gray', alpha=0.4)

# ticks
plt.xticks(color='white')
plt.yticks(color='white')

# spacing
plt.tight_layout()

plt.show()
```

```python
#Daily Price Change

# ensure datetime
df['Date'] = pd.to_datetime(df['Date'])
df = df.copy()

# daily change
df.loc[:, 'Gold_Change'] = df['Gold_24K_1g'].diff()

# dark theme
plt.style.use('dark_background')

plt.figure(figsize=(14,6))

# positive changes (green)
plt.plot(
    df['Date'],
    df['Gold_Change'],
    color='#00FF7F',
    linewidth=1.5,
    label='Daily Change'
)

# zero line 
plt.axhline(0, color='gray', linestyle='--', linewidth=1)

# fill positive/negative
plt.fill_between(
    df['Date'],
    df['Gold_Change'],
    where=(df['Gold_Change'] > 0),
    color='green',
    alpha=0.2
)

plt.fill_between(
    df['Date'],
    df['Gold_Change'],
    where=(df['Gold_Change'] < 0),
    color='red',
    alpha=0.2
)

# title & labels
plt.title("Daily Gold Price Change", fontsize=16, fontweight='bold', color='white')
plt.xlabel("Date", fontsize=12, color='white')
plt.ylabel("Change (INR)", fontsize=12, color='white')

# grid
plt.grid(True, linestyle='--', linewidth=0.5, color='gray', alpha=0.4)

# ticks
plt.xticks(color='white')
plt.yticks(color='white')

# legend
plt.legend(facecolor='black', edgecolor='white')

# spacing
plt.tight_layout()

plt.show()
```

```python
#Percentage Change

# ensure datetime
df['Date'] = pd.to_datetime(df['Date'])
df = df.copy()

# daily return %
df.loc[:, 'Gold_Return_%'] = df['Gold_24K_1g'].pct_change(fill_method=None) * 100

# dark theme
plt.style.use('dark_background')

plt.figure(figsize=(14,6))

# return line
plt.plot(
    df['Date'],
    df['Gold_Return_%'],
    color='#00FFFF',  # cyan
    linewidth=1.5,
    label='Daily Return (%)'
)

# zero line (important)
plt.axhline(0, color='gray', linestyle='--', linewidth=1)

# fill positive returns
plt.fill_between(
    df['Date'],
    df['Gold_Return_%'],
    where=(df['Gold_Return_%'] > 0),
    color='green',
    alpha=0.2
)

# fill negative returns
plt.fill_between(
    df['Date'],
    df['Gold_Return_%'],
    where=(df['Gold_Return_%'] < 0),
    color='red',
    alpha=0.2
)

# title & labels
plt.title("Gold Daily Return (%)", fontsize=16, fontweight='bold', color='white')
plt.xlabel("Date", fontsize=12, color='white')
plt.ylabel("Return (%)", fontsize=12, color='white')

# grid
plt.grid(True, linestyle='--', linewidth=0.5, color='gray', alpha=0.4)

# ticks
plt.xticks(color='white')
plt.yticks(color='white')

# legend
plt.legend(facecolor='black', edgecolor='white')

# spacing
plt.tight_layout()

plt.show()
```

```python
#Correlation
corr = df[['Gold_24K_1g', 'Silver_1g']].corr()
print(corr)
```

```python
#Monthly Trend

# ensure datetime
df['Date'] = pd.to_datetime(df['Date'])
df = df.copy()

# monthly data
df['Month'] = df['Date'].dt.to_period('M')
monthly_avg = df.groupby('Month')['Gold_24K_1g'].mean()

# convert Period → datetime (IMPORTANT for plotting)
monthly_avg.index = monthly_avg.index.to_timestamp()

# dark theme
plt.style.use('dark_background')

plt.figure(figsize=(14,6))

# main line
plt.plot(
    monthly_avg.index, 
    monthly_avg.values,
    color='#FFD700',
    linewidth=2.5,
    label='Monthly Avg Price'
)

# markers
plt.scatter(
    monthly_avg.index,
    monthly_avg.values,
    color='white',
    s=15,
    alpha=0.7
)

# area fill (premium look)
plt.fill_between(
    monthly_avg.index,
    monthly_avg.values,
    color='#FFD700',
    alpha=0.1
)

# title & labels
plt.title("Monthly Gold Price Trend", fontsize=16, fontweight='bold', color='white')
plt.xlabel("Month", fontsize=12, color='white')
plt.ylabel("Price (INR)", fontsize=12, color='white')

# grid
plt.grid(True, linestyle='--', linewidth=0.5, color='gray', alpha=0.4)

# ticks
plt.xticks(color='white', rotation=45)
plt.yticks(color='white')

# legend
plt.legend(facecolor='black', edgecolor='white')

# spacing
plt.tight_layout()

plt.show()
```

```python
#Volatility

# ensure datetime
df['Date'] = pd.to_datetime(df['Date'])
df = df.copy()

# volatility (30-day std)
df.loc[:, 'Volatility'] = df['Gold_24K_1g'].rolling(window=30).std()

# dark theme
plt.style.use('dark_background')

plt.figure(figsize=(14,6))

# main volatility line
plt.plot(
    df['Date'], 
    df['Volatility'], 
    color='#00FFFF',  # cyan
    linewidth=2,
    label='Volatility (30-Day)'
)

# area fill (premium)
plt.fill_between(
    df['Date'], 
    df['Volatility'], 
    color='#00FFFF', 
    alpha=0.1
)

# highlight peaks (high volatility)
high_vol = df['Volatility'] > df['Volatility'].mean()

plt.scatter(
    df['Date'][high_vol],
    df['Volatility'][high_vol],
    color='red',
    s=15,
    label='High Volatility'
)

# title & labels
plt.title("Gold Price Volatility (30-Day)", fontsize=16, fontweight='bold', color='white')
plt.xlabel("Date", fontsize=12, color='white')
plt.ylabel("Volatility", fontsize=12, color='white')

# grid
plt.grid(True, linestyle='--', linewidth=0.5, color='gray', alpha=0.4)

# ticks
plt.xticks(color='white')
plt.yticks(color='white')

# legend
plt.legend(facecolor='black', edgecolor='white')

# spacing
plt.tight_layout()

plt.show()
```
